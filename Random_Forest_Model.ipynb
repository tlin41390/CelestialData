```python
# Select 'feat_3' collection, and read in as pandas DataFrame
collection = db[FEAT_3_COLLECTION_NAME].find()
features_3_df = pd.DataFrame(list(collection))

# Drop mongo generated hash-code id column
features_3_df = features_3_df.drop(columns = ["_id"])
features_3_df
```

```python
# Select 'feat_30' collection, and read in as pandas DataFrame
collection = db[FEAT_30_COLLECTION_NAME].find()
features_30_df = pd.DataFrame(list(collection))

# Drop mongo generated hash-code id column
features_30_df = features_30_df.drop(columns = ["_id"])
features_30_df
```

```python
# Gets full path of current working directory 
cwd_bytes = subprocess.check_output("pwd")
cwd = cwd_bytes.decode("utf-8").rstrip("\n") + "/"
# print(cwd)
```

# Supervised Machine Learning Model "RandomForestClassifier" to Predict `label` from feature data

### Begin `features_3_sec.csv` section

```python
# Separate Features (X) from the Target (y)
y_3 = features_3_df["label"]
X_3 = features_3_df.drop(columns="label", axis=1)
```

```python
# Drop `filename` (identifier/same as `label`) and `length` (same for all songs - 3 sec sample)
# from features X
X_3 = X_3.drop(columns=["filename", "length"], axis=1)
X_3
```

```python
# Encode class labels in target y
genre_nums = {}
cnt = 0
for genre in y_3.unique():
    genre_nums[genre] = cnt
    cnt += 1

y_3 = y_3.apply(lambda x: genre_nums[x])
genre_nums
```

```python
# Split into training and testing sets
X_3_train, X_3_test, y_3_train, y_3_test = train_test_split(X_3, y_3, random_state=0)
```

```python
# Instantiate the model
rf_3_model = RandomForestClassifier(n_estimators=1400, min_samples_split=2, min_samples_leaf=1,
                                    max_features="auto", max_depth=40, bootstrap=False)  # CHANGEABLE for bar plot comparisons

# Fit the model
rf_3_model = rf_3_model.fit(X_3_train, y_3_train)

# Make predicitions using the testing data
rf_3_pred = rf_3_model.predict(X_3_test)
results = pd.DataFrame({
    "Prediction": rf_3_pred,
    "Actual": y_3_test
}).reset_index(drop=True)
results
```

```python
# Generate confusion matrix as colored heatmap
cm_3 = sklearn.metrics.confusion_matrix(y_3_test, rf_3_pred)

sortedlabels = genre_nums.keys()
cm_3_df = pd.DataFrame(cm_3, index=sortedlabels, columns=sortedlabels)

sns.set(font_scale=1.3)
sns.heatmap(cm_3_df, linewidths=0.5, cmap=sns.light_palette((1, 0.2, 0.6), n_colors=10000), annot=True)
sns.set(rc={"figure.figsize": (15.0, 13.0)})

plt.title("Confusiuon Matrix Heatmap for 3 Second Sample Features")
plt.xlabel("Predicted")
plt.ylabel("True")
```

```python
# Calculate Accuracy Score
acc_score_3 = sklearn.metrics.accuracy_score(y_3_test, rf_3_pred)
acc_score_3
```

```python
# Print classification report
print(sklearn.metrics.classification_report(y_3_test, rf_3_pred))
```

```python
classification_report_3_second = sklearn.metrics.classification_report(y_3_test, rf_3_pred, output_dict=True)
genre_metrics_3 = {}
for i in range(10):
    genre_metrics_3[str(i)] = classification_report_3_second[str(i)]
total_accuracy_3 = classification_report_3_second["accuracy"]
common_metrics_3 = {"macro avg": classification_report_3_second["macro avg"],
                  "weighted avg": classification_report_3_second["weighted avg"]}
genre_metrics_3_df = pd.DataFrame.from_dict(genre_metrics_3)
genre_metrics_3_df.columns = list(genre_nums.keys())
common_metrics_3_df = pd.DataFrame.from_dict(common_metrics_3)
```

```python
print("Total Accuracy:", total_accuracy_3)
```

```python
# Most important features
# List in descending order by feature importance
importances = rf_3_model.feature_importances_
sorted(zip(importances, X_3), reverse=True)[:5]
```
